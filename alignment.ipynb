```python
files = listdir('./wavs/day_5(04001-05000)')
print(len(files))
files[0]
```

```python
def clean_text(text):
    pattern = '(\.wav)$'
    text = re.sub(pattern=pattern, repl='', string=text)
    text = text.strip()
    text = " ".join(text.split())
    return text   
```

```python
wav_files = []

for i in files:
    j = "NC_Base/NcFemale/item/" + i
    k = clean_text(j)
    wav_files.append(k)
```
